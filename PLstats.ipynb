```python
PL['conv'] =( PL['Gls']/PL['xG'] ) * 100
PL['Contribution'] = PL['Gls'] + PL['Ast']
```

Picking players as per criterion

```python
CONVE = PL[~PL['Gls'].isna() & (PL['Contribution'] > 7) & (PL['Gls'] > 3)]
pd.set_option('display.precision', 2)
conversion = CONVE[['Player','conv','Team']]
conversion = conversion.sort_values(by=['conv'], ascending=False)
conversion.head(15)
```
